```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('2021_lloguer_preu_trim.csv')
df
```

```python
summary = df.describe()
print(summary)
```

```python
#  variable categorica (Districte or Barri)
df['Nom_Districte'].value_counts().plot(kind='bar')
plt.title('Distribució per Districte')
plt.xlabel('Districte')
plt.ylabel('Nombre de lloguers')
plt.show()
```

```python

df['Preu'] = pd.to_numeric(df['Preu'], errors='coerce')
df = df[df['Preu'] > 250] # eliminem outliers
df
```

```python

df['Preu'].plot(kind='hist')
plt.title('Distribució Preu')
plt.xlabel('Preu')
plt.ylabel('Nombre lloguers')
plt.show()
```

```python
df.boxplot(column='Preu', by='Nom_Districte')
plt.title('Distribució del Preu per Districte')
plt.xticks(rotation=90)
plt.xlabel('Districte')
plt.ylabel('Preu')
plt.show()
```

```python
# Any o Trimestre i Preu)
df.plot(x='Codi_Barri', y='Preu', kind='scatter')
plt.title('Preu per Codi_Barri')
plt.xlabel('Codi_Barri')
plt.ylabel('Preu')
plt.show()
```

```python
# 3 variables (Barri o Districte, Trimestre, i Preu)
df.pivot_table(index='Codi_Barri', columns='Trimestre', values='Preu').plot(kind='bar', stacked=True)
plt.title('Preu per Trimestre i Codi_Barri')
plt.xlabel('Codi_Barri')
plt.ylabel('Preu')
plt.show()
```
